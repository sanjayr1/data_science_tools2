```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
train_df.isnull().any()
```

```python
train_df.dtypes
```

```python
train_df.datetime = pd.to_datetime(train_df.datetime)
train_df.dtypes
```

```python
train_df.columns
```

```python
def engineered_time_feature(df):
    df['month'] = df.datetime.dt.month
    df['year'] = df.datetime.dt.year
    df['day'] = df.datetime.dt.day
    df['day_of_the_week'] = df.datetime.dt.day
    df['hour'] = df.datetime.dt.hour   
    
    
    return df
```

```python
test_df = pd.read_csv('test.csv')
```

```python
test_df.datetime= pd.to_datetime(test_df.datetime)
test_df.dtypes
```

```python
train_df = engineered_time_feature(train_df)
train_df['count'] = np.log1p(train_df['count'])
train_df.head()
```

```python
train_df.dtypes
```

```python
test_df = engineered_time_feature(test_df)
test_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set_df, val_set_df = train_test_split(train_df, test_size = .2, random_state= 1)
```
